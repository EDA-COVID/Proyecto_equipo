```python
import pandas as pd
import requests
import json
```

```python
def jsonlink_df(url_json):
    """
    What it does:
        # This function opens a json in python from a given url/link

    What it needs:
        # The url of the destination file, must be a string.

    What it returns:
        # The opened json as dataframe 
        GITHUB ID: @anaisvh
        """

    r = requests.get(url=url_json)
    json_readed = json.loads(r.text)
    df = pd.DataFrame(json_readed)
    return df
```

```python
# se debe importar a traves de la url, para asegurarnos de que los datos importados son actualizados
data_covid = jsonlink_df('https://covid.ourworldindata.org/data/owid-covid-data.json')
data_covid
```
